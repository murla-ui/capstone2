## Cleaning

```python
df.iloc[:, [1]].dtypes
```

```python
df = df.replace(',','.', regex=True)
```

```python
#getting rid from object datatype
for _ in range(0,26):
    if df.iloc[:, [_]].dtypes.dtype == 'object':
        df.iloc[:, [_]] = df.iloc[:, [_]].astype(float)
```
